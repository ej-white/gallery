```python
import numpy as np
import matplotlib.pyplot as plt

gr = 1.618
```

```python
def draw_triangle(x, h, x_min=0, x_max=100, y_max=100, zorder=1):
    # one line from  starting point to corner
    a_x1 = x_min; a_y1 = x
    a_x2 = x_max; a_y2 = y_max
    plt.plot([a_x1, a_x2], [a_y1, a_y2], 'k-', lw=1, zorder=zorder)

    # second line from  starting point to corner
    b_x1 = a_x1; b_y1 = a_y1 - h
    b_x2 = a_x2; b_y2 = a_y2
    plt.plot([b_x1, b_x2], [b_y1, b_y2], 'k-', lw=1, zorder=zorder)

    # vertical line between each line's starting point
    c_x1 = a_x1; c_y1 = a_y1
    c_x2 = b_x1; c_y2 = b_y1
    plt.plot([c_x1, c_x2], [c_y1, c_y2], 'k-', lw=1, zorder=zorder)
```

```python
plt.rcParams["figure.figsize"] = [14, 10]
plt.rcParams["figure.autolayout"] = False

xmin = 0
xmax = 140
ymin = 0
ymax = 100

# Set the axis
figure, axes = plt.subplots()
axes.set_xlim((xmin - 1, xmax + 1))
axes.set_ylim((ymin - 1, ymax + 1))

spacing = 5
for i in range(100, 0, -spacing):
    draw_triangle(i, spacing/2, x_min=xmin, x_max=xmax, y_max=ymax, zorder=1)
    draw_triangle(i-spacing/2, spacing/2, x_min=xmax, x_max=xmin, y_max=ymin, zorder=1)

# Final touches
plt.axis('off')

radius = 24
cc_fill = plt.Circle((xmax/2, ymax/2), radius, color='white', fill=True, zorder=2)
axes.add_artist(cc_fill)
cc_border = plt.Circle((xmax/2, ymax/2), radius, color='k', fill=False, zorder=3)
axes.add_artist(cc_border)

plt.show()
```

```python
figure.savefig("triangles.svg", format='svg', dpi=1200)  # dpi doesn't make a difference?
```
